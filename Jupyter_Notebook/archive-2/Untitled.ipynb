```python
import numpy as np
from scipy.sparse import csr_matrix
```

```python
dense_matrix = np.array([[1, 0, 0], [0, 0, 2], [3, 0, 4]])

# Convert the dense matrix to a CSR sparse matrix
sparse_matrix = csr_matrix(dense_matrix)

# Print the sparse matrix
print("CSR Sparse Matrix:")
print(sparse_matrix)
```

```python
print("\nElement at (0, 0):", sparse_matrix[0, 0])
print("Element at (2, 0):", sparse_matrix[2, 0])
```

```python
from scipy.sparse import coo_matrix

# Create a dense matrix
dense_matrix = np.array([[1, 0, 0], [0, 0, 2], [3, 0, 4]])

# Convert the dense matrix to a COO sparse matrix
rows, cols = np.nonzero(dense_matrix)  # Get the row and column indices of non-zero elements
values = dense_matrix[rows, cols]  # Get the non-zero values
sparse_matrix = coo_matrix((values, (rows, cols)), shape=dense_matrix.shape)

# Print the COO sparse matrix
print("COO Sparse Matrix:")
print(sparse_matrix)
```

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify=cancer.target, random_state=66)
```

```python
train_acc=[]
test_acc=[]

neighbors=range(1,11)

for n_neighbors in neighbors:
    clf=KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(X_train,y_train)
    train_acc.append(clf.score(X_train,y_train))
    test_acc.append(clf.score(X_test,y_test))
```

```python
train_acc
```

```python
plt.plot(neighbors, train_acc, label="training accuracy")
plt.plot(neighbors, test_acc, label="test accuracy")
plt.ylabel("Accuracy")
plt.xlabel("n_neighbors")
plt.legend()
```

```python
from sklearn.neighbors import KNeighborsRegressor 
X, y = mglearn.datasets.make_wave(n_samples=40)
# split the wave dataset into a training and a test set
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
# instantiate the model and set the number of neighbors to consider to 3
reg = KNeighborsRegressor(n_neighbors=3)
# fit the model using the training data and training targets 
reg.fit(X_train, y_train)
```

```python
print(reg.predict(X_test))
```

```python
print(reg.score(X_test, y_test))
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
# create 1,000 data points, evenly spaced between -3 and 3 
line = np.linspace(-3, 3, 1000).reshape(-1, 1)
for n_neighbors, ax in zip([1, 3, 9], axes):
    # make predictions using 1, 3, or 9 neighbors
    reg = KNeighborsRegressor(n_neighbors=n_neighbors)
    reg.fit(X_train, y_train)
    ax.plot(line, reg.predict(line))
    ax.plot(X_train, y_train, '^', c=mglearn.cm2(0), markersize=8)
    ax.plot(X_test, y_test, 'v', c=mglearn.cm2(1), markersize=8)
    ax.set_title(
    "{} neighbor(s)\n train score: {:.2f} test score: {:.2f}".format(
        n_neighbors, reg.score(X_train, y_train),
        reg.score(X_test, y_test)))
    ax.set_xlabel("Feature")
    ax.set_ylabel("Target")
axes[0].legend(["Model predictions", "Training data/target",
                    "Test data/target"], loc="best")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
# Load the Breast Cancer dataset
data = load_breast_cancer()
X = data.data  # Features
y = data.target  # Target labels
```

```python
# Select the "texture error" feature (change index as needed)
feature_index = data.feature_names.tolist().index("texture error")
selected_feature = X[:, feature_index]
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(selected_feature.reshape(-1, 1), y, test_size=0.2, random_state=42)

# Train a logistic regression model
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# Plot the data points and the logistic regression decision boundary
plt.figure(figsize=(8, 6))

# Plot data points
plt.scatter(X_test, y_test, color='black', label='Actual Data')

# Plot decision boundary
X_range = np.linspace(selected_feature.min(), selected_feature.max(), 300).reshape(-1, 1)
decision_boundary = model.predict_proba(X_range)[:, 1]
plt.plot(X_range, decision_boundary, color='red', linewidth=3, label='Logistic Regression Decision Boundary')
```

```python
# Highlight points predicted as class 1
class_1_points = X_test[model.predict(X_test) == 1]
plt.scatter(class_1_points, np.ones_like(class_1_points), color='red', marker='^', label='Predicted Class 1')

# Set labels and title
plt.xlabel("Texture Error")
plt.ylabel("Target Label")
plt.title("Relationship Between Texture Error and Target Label")
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

# Load the Breast Cancer dataset
data = load_breast_cancer()
X = data.data  # Features
y = data.target  # Target labels (0 for malignant, 1 for benign)

# Select the "texture error" feature (change index as needed)
feature_index = data.feature_names.tolist().index("texture error")
selected_feature = X[:, feature_index]

# Create a scatter plot
plt.figure(figsize=(8, 6))

# Scatter plot for malignant tumors (class 0)
plt.scatter(selected_feature[y == 0], y[y == 0], color='red', label='Malignant')

# Scatter plot for benign tumors (class 1)
plt.scatter(selected_feature[y == 1], y[y == 1], color='green', label='Benign')

# Set labels and title
plt.xlabel("Texture Error")
plt.ylabel("Tumor Type (0: Malignant, 1: Benign)")
plt.title("Relationship Between Texture Error and Tumor Type")

# Show legend
plt.legend()

# Show the plot
plt.show()
```

```python
feature_names = data.feature_names

# Create a grid of subplots
fig, axes = plt.subplots(nrows=5, ncols=6, figsize=(18, 15))

# Flatten the axes for easier indexing
axes = axes.flatten()

# Iterate through features and plot for each label
for i, feature in enumerate(feature_names):
    # Select the feature
    selected_feature = X[:, i]

    # Create a scatter plot for malignant tumors (class 0)
    axes[i].scatter(selected_feature[y == 0], y[y == 0], color='red', label='Malignant', alpha=0.7)

    # Create a scatter plot for benign tumors (class 1)
    axes[i].scatter(selected_feature[y == 1], y[y == 1], color='green', label='Benign', alpha=0.7)

    # Set labels and title
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel("Tumor Type (0: Malignant, 1: Benign)")
    axes[i].set_title(f"{feature} vs Tumor Type")

# Adjust layout
plt.tight_layout()

# Show legend in the last subplot
axes[-1].legend()

# Show the plot
plt.show()
```
