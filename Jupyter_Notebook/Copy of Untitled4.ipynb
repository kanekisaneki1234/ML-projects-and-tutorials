```python
import pandas as pd
import requests

# Define the URL of the web page
url = 'https://en.wikipedia.org/wiki/2014_in_anime'

# Send an HTTP GET request to the URL and store the response
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    html_content = response.text
else:
    print('Failed to retrieve the web page.')
```

```python
# Parse HTML tables from the HTML content
tables = pd.read_html(html_content)

# Check if any tables were found on the web page
if tables:
    # Assuming you want to work with the first table found (index 0)
    df = tables[0]
else:
    print('No tables found on the web page.')
```
