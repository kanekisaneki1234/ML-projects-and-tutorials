```python
import pandas as pd
import random

# Create lists for age groups, genders, and music genres
# age_groups = ['18-24', '25-34', '35-44', '45-54', '55+']
genders = ['Male', 'Female']
music_genres = ['Pop', 'Rock', 'Hip-Hop', 'Jazz', 'Country', 'Electronic', 'Classical', 'R&B']

# Generate random data for the dataset
data = []
for _ in range(30):  # Generate data for 100 individuals
    age = random.randint(20,41)
    gender = random.choice(genders)
    genre = random.choice(music_genres)
    data.append([age, gender, genre])

# Create a DataFrame from the generated data
columns = ['Age Group', 'Gender', 'Music Genre']
df = pd.DataFrame(data, columns=columns)

# Display the first few rows of the dataset
print(df.head())
```
